```python
import matplotlib.pyplot as plt
import pandas as pd


#I. Plot the sepal length of the Iris dataset against the corresponding index values

df = pd.read_csv("D:\Downloads\Iris.csv")
plt.figure(figsize=(10, 6))
plt.plot(df.index,df['Sepal_Length'],'-o')
plt.title("Sepal Length of User's Iris Dataset")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")

```

```python
#II . Create a scatter plot to plot the sepal length and width of the Iris dataset against each other

plt.figure(figsize=(10, 6))
plt.scatter(df['Sepal_Length'],df['Sepal_Width'],color = 'purple')
plt.title("Sepal Length vs Sepal Width of Iris Dataset")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")

```

```python
# Using a bar graph plot the number of flowers of each species in the Iris dataset.

Species = df['Species'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(Species.index,Species.values,color=['blue', 'green', 'red'])
plt.title("Number of Flowers for Each Species in the Iris Dataset")
plt.xlabel("Species")
plt.ylabel("Count")
plt.xticks(rotation=45)
```
